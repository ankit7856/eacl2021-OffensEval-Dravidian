### Imports

```
import jsonlines
import os
from collections import Counter
from sklearn.metrics import classification_report
import pandas as pd
import numpy as np
```

### Doing the ensemble (dev)

```
mode = "dev"
for lang_ckpts in [kannada_ckpts,]: # malayalam_ckpts, tamil_ckpts]:
    print("\n\n")
    predictions_ensemble = None
    truth_ensemble = None
    for ckpt in lang_ckpts[mode]:
        filename = os.path.join(ckpt, "predictions.jsonl")
        lines = [line for line in jsonlines.open(filename)]
        print(len(lines))
        predictions = [line["prediction"] for line in lines]
        truth = [line["label"] for line in lines]
        print(classification_report(truth, predictions, digits=4))

        if predictions_ensemble is None:
            predictions_ensemble = [[] for _ in lines]
        if truth_ensemble is None:
            truth_ensemble = [line["label"] for line in lines]
        for i, line in enumerate(lines):
            predictions_ensemble[i].append(line["prediction"])

    print("**********************")
    n_ensemble = len(predictions_ensemble[0])
    print(n_ensemble)
    # pred_labels = [max(lst, key=lst.count) for lst in results_ensemble]
    pred_labels = [Counter(lst).most_common(1)[0][0] for lst in predictions_ensemble]
    report = classification_report(truth_ensemble, pred_labels, digits=4)
    print("\n" + report)
    print("******* END *******")
```
